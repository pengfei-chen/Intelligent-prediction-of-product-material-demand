### 第一步，按照 df_stock 的日期，对 df 进行切分，然后再汇总

```python
df_copy = df.copy()
```

```python
df_copy['过账日期'] = df_copy['过账日期'].map(lambda x:x[:10])
```

```python
df_1 = pd.DataFrame()
for data in df_stock['过账日期'].unique():
    df_s = df_copy[df_copy['过账日期'] <= data]
    df_copy = df_copy[df_copy['过账日期'] > data]
    df_s['过账日期_汇总'] = data
    df_1 = pd.concat([df_1, df_s])

df_1.shape
```

```python
df_2 = df_1.groupby(['工厂编码','物料编码','物料品牌','物料类型','物料品类','过账日期_汇总'])['需求量'].agg('sum').reset_index()
df_2.head()
```

```python
df_stock.rename(columns={'过账日期':'过账日期_汇总'}, inplace=True)
```

```python
df_2.shape, df_stock.shape
```

```python
df_3 = pd.merge(df_2, df_stock, how ='left')
df_3
```

**排序**

```python
df_3 = df_3.sort_values(['工厂编码','物料编码','物料品牌','物料类型','物料品类','过账日期_汇总'])
df_3.head(3)
```

```python
df_3.head(3).append(df_3.tail(3))
```

**库存量有缺失值.**

```python
df_3['库存量'] = df_3['库存量'].fillna(0)
```

```python
df_3['上月库存余量'] = df_3.groupby(['工厂编码','物料编码','物料品牌','物料类型','物料品类']).apply(lambda x\
                                                         :x['库存量'].shift(1) ).reset_index(drop=True).fillna(0)
df_3
```

```python
df_3['当月实际需求量'] = df_3['需求量'] + df_3['上月库存余量'] - df_3['库存量']
df_3
```

## 接下来，我以预测“当月实际需求量”为目标。 预测完之后，当月实际需求量 + 库存量 - 上月库存余量， 就可以得到 需求量！

```python
from sklearn.model_selection import cross_val_score
import lightgbm as lgb

"""定义优化函数"""
def rf_lgb(num_leaves, max_depth, bagging_fraction, feature_fraction, bagging_freq, min_data_in_leaf, 
              min_child_weight, min_split_gain, reg_lambda, reg_alpha):
    # 建立模型
    model_lgb = lgb.LGBMRegressor(boosting_type='gbdt', bjective='mape', metric='mean_absolute_error',
                                   learning_rate=0.1, n_estimators=5000,
                                   num_leaves=int(num_leaves), max_depth=int(max_depth), 
                                   bagging_fraction=round(bagging_fraction, 2), feature_fraction=round(feature_fraction, 2),
                                   bagging_freq=int(bagging_freq), min_data_in_leaf=int(min_data_in_leaf),
                                   min_child_weight=min_child_weight, min_split_gain=min_split_gain,
                                   reg_lambda=reg_lambda, reg_alpha=reg_alpha,
                                   n_jobs= 8
                                  )
    
    val = cross_val_score(model_lgb, train_x, train_y, cv=3, scoring = 'neg_mean_absolute_percentage_error').mean()
    
    return val
```
